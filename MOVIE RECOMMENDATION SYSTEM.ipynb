## DATA PRE-PROCESSING AND CLEANING

### Renaming and Merging Columns

```python
# Rename the 'movie_id' column in the 'credits' DataFrame to 'id'
credits_column_renamed = df2.rename(index=str, columns={"movie_id": "id"})

# Merge the 'movies' and 'credits_column_renamed' DataFrames based on the 'id' column
movies_merge = df1.merge(credits_column_renamed, on='id')

# Print the first few rows of the merged DataFrame
movies_merge.head()
```

### Drop specified columns

```python
# Drop specified columns from the 'movies_merge' DataFrame and create a new DataFrame 'movies_cleaned'
movies_cleaned = movies_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
```

### Replace any non-standard NaN representations with NumPy NaN

```python
# Replace any non-standard NaN representations with NumPy NaN
movies_cleaned['overview'] = movies_cleaned['overview'].replace('your_non_standard_representation', np.nan)
```

## Feature Extraction

### TF-IDF (Term Frequency-Inverse Document Frequency)

```python
# Import necessary libraries
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Replace any remaining NaN values in the 'overview' column with an empty string
movies_cleaned['overview'] = movies_cleaned['overview'].replace(np.nan, '')

# Create an instance of TfidfVectorizer
tfv = TfidfVectorizer(
    min_df=3,                    # Ignore terms that appear in less than 3 documents
    max_features=None,           # Keep all features (no limit on the number of features)
    strip_accents='unicode',     # Remove accents during the preprocessing step
    analyzer='word',             # Analyze the data at the word level
    token_pattern=r'\w{1,}',      # Use words that consist of at least one alphanumeric character
    ngram_range=(1, 3),           # Consider unigrams, bigrams, and trigrams
    stop_words='english'         # Exclude common English stop words
)

# Fit and transform the 'overview' text after handling missing values
tfv_matrix = tfv.fit_transform(movies_cleaned['overview'])
```

```python
# Print the TF-IDF matrix
print(tfv_matrix)
```

```python
# Print the shape of the TF-IDF matrix
print(tfv_matrix.shape)
```

 #### There are 4803 rows (each corresponding to a document) and 10417 columns (each corresponding to a unique term in the entire collection).

### Computes the sigmoid kernel between two matrices

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
print(sig[0])
```

### Finding movies that are most similar to 'Newlyweds' based on the computed sigmoid kernel values

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_cleaned.index, index=movies_cleaned['original_title']).drop_duplicates()

# Print the indices Series
print(indices)

# Get the index of the movie 'Newlyweds'
JohnCarter_index = indices['John Carter']
print(JohnCarter_index)

# Print the sigmoid kernel values for the movie 'Newlyweds'
print(sig[JohnCarter_index])

# Enumerate and print the similarity scores for the movie 'Newlyweds' with all other movies
enumerate_scores = list(enumerate(sig[JohnCarter_index]))

# Sort the similarity scores in descending order
sorted_scores = sorted(enumerate_scores, key=lambda x: x[1], reverse=True)

# Print the sorted similarity scores
print(sorted_scores)
```

### Top 10 recommendations for movies

```python
def give_recommendations(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwise similarity scores
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies based on similarity scores
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies (excluding the input movie itself)
    sig_scores = sig_scores[1:11]

    # Get the indices of the top 10 most similar movies
    movie_indices = [i[0] for i in sig_scores]

    # Return the titles of the top 10 most similar movies
    return movies_cleaned['original_title'].iloc[movie_indices]
```

```python
print(give_recomendations('The Dark Knight Rises'))
```
